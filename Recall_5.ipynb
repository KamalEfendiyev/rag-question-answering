```python
def recall_at_k(search_fn, k=5, num_samples=100):
    correct = 0

    for i in range(num_samples):
        query = queries[i]
        true_context = dataset[i]['context']

        results = search_fn(query, top_k=k)

        if true_context in results:
            correct += 1

    return correct / num_samples

bm25_recall = recall_at_k(bm25_search)
dense_recall = recall_at_k(dense_search)

print("BM25 Recall@5:", bm25_recall)
print("Dense Recall@5:", dense_recall)
```
